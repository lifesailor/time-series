# 1. Time Series Components

1. Systematic: Components of the time series that have consistency or reoccurence and can be described and modeled.

2. Non-systematic: Components of time series that cannot be directly modeled.

Time series decomposition involves thinking of a series as a combination of level, trend, seasonality, and noise components.

1. level: The average value in the series
2. trend: The increasing or decreasing value in the series.
3. seasonality: The repeating short-term cycle in the series.
4. noise: The random variation in the series.

# 2. Combining time series components

### 1. Additive

- y(t) = Level + Trend + Seasonality + Noise

An additive model is linear where changes over time are consistently made by the same amount. A linear tredn is a straight line. A linear seasonality has the same frequency

### 2. Multiplicative Model

- y(t) = Level × Trend × Seasonality × Noise

A multiplicative model is nonlinear, such as quadratic or exponential. Changes increase or decrease over time. A nonlinear trend is a curved line. A nonlinear seasonality has an increasing or decreasing frequency and/or amplitude over time.

# 3. Automatic Time Series Decomposition

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### 1. Additive Decomposition

```python
from random import randrange
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose

series = [i + randrange(10) for i in range(100)]
result = seasonal_decompose(series, model='additive', freq=1)
result.plot()
pyplot.show()
```

### 2. Multiplicative Decomposition

```python
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', freq=1) 
result.plot()
pyplot.show()
```

### 3. Airline Passengers Dataset

```python
from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose

series = Series.from_csv('./AirPassengers.csv', header=0)
result = seasonal_decompose(series, model='multiplicative')
result.plot()
pyplot.show()
```

언제 시계열 분해법을 additive or multiplicative로 사용할 것인가.

### 1. multiplicative

In a multiplicative time series, the components multiply together to make the time series. If you have an increasing trend, the amplitude of seasonal activity increases. Everything becomes more exaggerated. This is common when you’re looking at web traffic.

### 2. additive

In an additive time series, the components add together to make the time series. If you have an increasing trend, you still see roughly the same size peaks and troughs throughout the time series. This is often seen in indexed time series where the absolute value is growing but changes stay relative.
