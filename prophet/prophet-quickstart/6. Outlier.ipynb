# 1. Package and Data Load

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
df = pd.read_csv('../prophet-quickstart/prophet/examples/example_wp_R_outliers1.csv')
```

# 2. Outlier

```python
df['y'] = np.log(df['y'])
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(periods=1096)
forecast = m.predict(future)
m.plot(forecast)
```

# 3. Remove Outlier

```python
df.loc[(df['ds'] > '2010-01-01') & (df['ds'] < '2011-01-01'), 'y'] = None
model = Prophet().fit(df)
```

```python
forecast2 = model.predict(future)
```

```python
m.plot(forecast2)
```
