# 1. Data and Package Load

```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
df = pd.read_csv('./prophet/examples/example_wp_peyton_manning.csv')
```

```python
df['y'] = np.log(df['y'])
```

```python
df.head()
```

```python
m = Prophet()
```

# 2. Fit

```python
m = Prophet()
m.fit(df)
```

# 3. Cross Validation

```python
from fbprophet.diagnostics import cross_validation
df_cv = cross_validation(m, horizon = '360 days')
df_cv.head()
```
